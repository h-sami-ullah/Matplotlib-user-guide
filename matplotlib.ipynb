**matplotlib**

Collection of functions that it function like MATLAB. This is the data visualization and graphical plotting library for Python and its numerical extension NumPy. In a nut shell it is basicaly a graph plotting library AND if we use it with Numpy then it is an Alternative to MATLAB.

It famous module is **pyplot** (to plot charts) that contains different functions like:


*   plot()
*   show()
*   scatter()
*   yscale()
*   xlabel()
*   ylabel()
*  title() 
*  xticks()
*   yticks()











**Syntax:** plt.plot(x-axis, y-axis, 'formating_style')

 The default format string is 'b-', which is a solid blue line.

```python
import matplotlib.pyplot as plt
apl_price =[74988,949234,8438,2498234,92489,872349,7382]
ms_price =[1458,54434,4438,2483234,9239,42349,2382]
year = [2011,2012,2013,2014,2015,2016,2017]

plt.plot(year, apl_price, '<-r')
plt.scatter(year, ms_price)
plt.ylabel("Prices--->")
plt.xlabel("Year--->")
plt.title('Stock Prices of cell 2011-17')
plt.xticks([2011,2017], ['year-2011','year-2017'])
plt.yscale('log')
plt.show()
```

**Strig Formating Styles:**

'-'       solid line style

'--'      dashed line style

'-.'      dash-dot line style

':'       dotted line style

'.'       point marker

','       pixel marker

'o'       circle marker

'v'       triangle_down marker

'^'       triangle_up marker

'<'       triangle_left marker

'>'       triangle_right marker

'1'       tri_down marker

'2'       tri_up marker

'3'       tri_left marker

'4'       tri_right marker

's'       square marker

'p'       pentagon marker


'*'       star marker

'h'       hexagon1 marker

'H'       hexagon2 marker


'+'       plus marker

'x'       x marker

'D'       diamond marker

'd'       thin_diamond marker

'|'       vline marker

'_'       hline marker

```python
import numpy as np

# evenly sampled time at 200 intervals
t = np.arange(0, 5, 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', 
         t, t**2, 'bs', 
         t, t**3, 'g^')
plt.show()
```
